```python
##Popularity Based Reccomendations System
```

```python
users=user.merge(anime,on='anime_id')
```

```python
num_ratings_df=users.groupby('Anime Title').count()['rating'].reset_index()
num_ratings_df.rename(columns={'rating':'num_rating'},inplace=True)
```

```python
num_ratings_df
```

```python
#avg_ratings=users.groupby('Anime Title').mean()['rating'].reset_index()
avg_ratings_df = users.groupby('Anime Title')['rating'].agg('mean').reset_index()
avg_ratings_df.rename(columns={'rating':'avg_rating'},inplace=True)
avg_ratings_df
```

```python
popular_df=num_ratings_df.merge(avg_ratings_df,on='Anime Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False)
```

```python
popular_df
```

```python
anime.rename(columns={'Name': 'Anime Title'}, inplace=True)
anime
```

```python
popular_df=popular_df.merge(anime,on='Anime Title').drop_duplicates('Anime Title')[['Anime Title','Studios','Image URL','num_rating','avg_rating']]
```

```python
popular_df
```

```python
##Collaborative Filtering Based Reccommender System
```

```python
users
```

```python
x=users.groupby('user_id').count()['rating']>100
weebs=x[x].index
```

```python
filtered=users[users['user_id'].isin(weebs)]
```

```python
y=filtered.groupby('Anime Title').count()['rating']>=100
famous_ani=y[y].index
```

```python
final_filtered=filtered[filtered['Anime Title'].isin(famous_ani)]
```

```python
final_filtered
```

```python
pt=final_filtered.pivot_table(index='Anime Title',columns='user_id',values='rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(pt)
```

```python
def reccomend(anime_name):
    index=np.where(pt.index==anime_name)[0][0]
    similar_items=sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:20]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df=anime[anime['Anime Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Anime Title')['Anime Title'].values))
        item.extend(list(temp_df.drop_duplicates('Anime Title')['Studios'].values))
        item.extend(list(temp_df.drop_duplicates('Anime Title')['Image URL'].values))

        data.append(item)
        
    return data
```

```python
reccomend('One Piece')
```
